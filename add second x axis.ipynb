```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)

def add_2nd_x_axis(ax, x2_tick_location_in_x):
    ''' add second x axis simply by add second axes with same xlim and ylim and sharey()'''
    ax2 = ax.twiny()
    x1, x2 = ax.get_xlim()
    y1, y2 = ax.get_ylim()
    ax2.set_xlim(x1, x2)
    ax2.set_ylim(y1, y2)

    def tick_function(X):
        V = X + 25
        return ['{0:.0f}'.format(x) for x in V]

    # set ax2 x-tick locations in the original x axis, then set the x-ticklabels
    ax2.set_xticks(x2_tick_location_in_x)
    ax2.set_xticklabels(tick_function(x2_tick_location_in_x))
    ax2.set_xlabel("x+25")
    return

# NOTE: add second x axis AFTER all other modifies of x limit (such as ax.plot or ax.set_xlim)
x2_tick_location_in_x = np.linspace(2, 10, 5)
add_2nd_x_axis(ax, x2_tick_location_in_x)
```
