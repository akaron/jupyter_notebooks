purpose of this notebook is to show how to sort legends manually.
To use it, copy the function get_sorted_handles_labels() and use it as demonstrated.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='sinx')
ax.plot(x, np.sin(2*x), label='sin2x')
ax.plot(x, np.sin(3*x), label='sin3x')

# if for each ax.plot save the returned boject (such as l1 = ax.plot(); l2=ax.plot(); ...)
# then one may sort this way:
# ax.legend([l2, l1, l3], ["cos(x)", "sin(x)", "sin(2x)"])

# sort
def get_sorted_handles_labels(ax, order):
    ''' ax: an axes object; order: a list/tuple of integers'''
    handles, labels = ax.get_legend_handles_labels()
    def get_list_order(listt, order):
        return [listt[i] for i in order]
    return get_list_order(handles, order), get_list_order(labels, order)

handles_sorted, labels_sorted = get_sorted_handles_labels(ax=ax, order=(2, 1, 0))
ax.legend(handles_sorted, labels_sorted, loc='lower right', frameon=True, fontsize=20)
plt.show()
```
