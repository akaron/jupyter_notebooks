```python
import matplotlib.pyplot as plt
```

```python
plt.rc('text', usetex=False)
plt.scatter((1, 2, 3), (0, 0, 1), marker=r"$\oplus$", s=100)
plt.text(2, 0.5, r'Earth $\oplus$', color='black', size='13')
plt.show()
```

# usetex=True
so one can use symbols from other latex packages

```python
plt.rc('text', usetex=True)
plt.rcParams['text.latex.preamble'] = [r"\usepackage{mathabx}",]
```

```python
plt.scatter((1, 2, 3), (1, 1, 1), marker=r"$\Earth$", s=100, label="Earth")
plt.scatter((1, 2, 3), (0, 0, 0), marker=r"$\oplus$", s=100, label="oplus")
plt.text(1.3, 0.6, r'Earth $\Earth$', color='black', size='13')
plt.text(1.3, 0.2, r'oplus $\oplus$', color='black', size='13')
# plt.legend(frameon=True)
plt.show()
```
